### Ajouter des données à nos graphiques

```python
x_values = data.iloc[0:12,0] 
y_values = data.iloc[0:12,1] 
plt.plot(x_values, y_values)
plt.show()
```

### Créer une rotation sur les données de l'axe x

```python
#plt.xticks()
```

```python
#matplotlib.pyplot.xticks(*args, **kwargs)
```

```python
#plt.xticks(rotation = 45)

#plt.yticks()
```

```python
x_values = data.iloc[0:12,0] 
y_values = data.iloc[0:12,1] 
plt.plot(x_values, y_values)
plt.xticks(rotation = 90)
#on fait une rotation de 90 degrés sur nos légendes 
plt.show()
```

### Ajouter les intitulés des axes et un titre à notre graphique

```python
#plt.xlabel()
#plt.ylabel()
#plt.title()
```

```python
plt.plot(x_values, y_values)
plt.xlabel("Taux de chomage(%)")
plt.ylabel("Mois")
plt.title("Evolution de taux de chomage en 1948 sur USA")
plt.xticks(rotation = 90)
plt.show()
```
