### Comparaison de graphiques

```python
fig = plt.figure(figsize=(12,12)) #on cree notre figure
for i in range(5): #on parcours 5 valeurs 
    ax=fig.add_subplot(5,1,i+1) #on cree un plot a chaque itération
    start_index= i*12 
    end_index = (i+1)*12
    subset = data[start_index:end_index]
    ax.plot(subset.loc[:,'DATE'],subset.loc[:,'UNRATE'])
plt.show()
```

### Superposition des courbes

```python
#pandas.Series.dt
data['MONTH']=data.loc[:,'DATE'].dt.month 
#Transforme date en mois
data['MONTH']
```

```python
#plt.plot(data.loc[0:12,"MONTH"],data.loc[0:12,"UNRATE"])
#plt.plot(data.loc[12:24,"MONTH"],data.loc[12:24,"UNRATE"])
```

```python
fig = plt.figure(figsize=(6,6))
plt.plot(data[0:12]["MONTH"],data[0:12]["UNRATE"])
plt.plot(data[12:24]["MONTH"],data[12:24]["UNRATE"], c="red")
#parametre c pour définir la couleur 
plt.show()
#cette syntaxe est importante pour que les deux courbes ne soient pas collées 
```

```python
fig = plt.figure(figsize=(6,3))
plt.plot(data[0:12]["MONTH"],data[0:12]["UNRATE"],c="red")
plt.plot(data[12:24]["MONTH"],data[12:24]["UNRATE"], c="blue")
plt.show()
```

### Ajouter plus de courbes

```python
fig = plt.figure(figsize=(10,6))
colors = ["red","blue","green","orange","black"]
i=1
for i in colors:
    c=colors.index(i)
    plt.plot(data[c*12:(c+1)*12]["MONTH"],data[c*12:(c+1)*12]["UNRATE"],c=i)

plt.show()
```

```python
fig = plt.figure(figsize=(10,6))
colors = ["red","blue","green","orange","black"]
i=1
for i in colors:
    c=colors.index(i)
    plt.plot(data[c*12:(c+1)*12]["MONTH"],data[c*12:(c+1)*12]["UNRATE"],c=i,label=1948+c)
    plt.legend(loc="upper left") 

plt.show()
```
