### Utiliser des index personnalisés

```python
#DataFrame.set_index(Series, inplace=True, drop=False)
#drop => Booléen. La valeur par défaut est “True” qui supprime la colonne à définir comme index
#inplace => Booléen. Si True, modifiez l’appelant DataFrame sur place
#Series => colonne ou liste de colonnes à définir comme index
```

```python
#on souhaite conserver la colonne FILM
#On souhaite garder le DataFrame d'origine et en créer un nouveau que 
#nous assignerons à la variable fandango_films
fandango_films = fandango.set_index(fandango["FILM"],inplace=False, drop=False)
```

```python
fandango_films.index
```

### Méthode Apply() sur les colonnes d'un DataFrame

```python
#apply()
#datafrmae.apply(lambda x: x**2)
```

```python
import numpy as np
```

```python
# retourner les types de donnees en tant que series pour toutes les colonnes
types =fandango_films.dtypes
types
```

```python
#filtrer les nombres à virgules float64 
#et retourner le nom des colonnes
float_columns = types[types.values == 'float64'].index
float_columns
```

```python
#utilisation des crochets pour filtrer 
#les colonnes de la condition float_columns
float_df = fandango_films[float_columns]
float_df
```

```python
#On calcule l'ecart-type de chaque colonne
deviations = float_df.apply(lambda x: np.std(x))
deviations
```
